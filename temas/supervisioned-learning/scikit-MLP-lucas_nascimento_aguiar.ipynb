```python
import pandas as pd

RANDOM_STATE = 11

df = pd.read_csv(
    "https://vincentarelbundock.github.io/Rdatasets/csv/MASS/biopsy.csv", index_col=0
)
df.head()

print("First 5 records:", df.head())
print("Data description:", df.describe())
```

```python
from sklearn.model_selection import train_test_split

# features = df[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9']]
features = df[["V1", "V2", "V3"]]
target = df[["class"]]
features_train, features_test, target_train, target_test = train_test_split(
    features, target, train_size=0.7, test_size=0.3, random_state=RANDOM_STATE
)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

inputer = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=0)

features_train = inputer.fit_transform(features_train)
features_test = inputer.fit_transform(features_test)
target_train = inputer.fit_transform(target_train)
target_test = inputer.fit_transform(target_test)

print("Shape of training data:", features_train.shape)
print("Shape of target data:", target_train.shape)
```
